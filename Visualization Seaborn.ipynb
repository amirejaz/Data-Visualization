# Categorical Categorical Plots

```python
#for heatmaps 
# using different data set
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = flights.pivot_table(index = 'year', columns='month', values='passengers', aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

# Clustermap
<!-- it recombine the similar rows and columns together -->

```python
#it combines the similar rows and columns together
```

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, z_score=0, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, metric='correlation')
```

```python
# if you don't want to cluster on the basis of rows then set (row_cluster = False)
sns.clustermap(x, z_score=0, annot=True, row_cluster=False, metric='correlation')
```

```python
# if you don't want to cluster on the basis of columns then set (col_cluster = False)
sns.clustermap(x, z_score=0, annot=True, col_cluster=False, metric='correlation')
```

# side by side grapghs

```python
# for jointplot i am using tips dataset
sns.jointplot(x='total_bill', y='tip', kind='scatter', data=tips) #by default it is scatter
#histogram side by side
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=tips) 
#hex is basically a 2d histogram
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)

#kde is basically a 2d PDF (probabilty density function)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
#it is a regression plot and used for prediction
```

# Distplot

```python
# using titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.distplot(titanic['age'])
```

```python
# only histogram
sns.distplot(titanic['age'], kde=False)
```

```python
#only kde
sns.distplot(titanic['age'], hist=False)
```

```python
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
```
